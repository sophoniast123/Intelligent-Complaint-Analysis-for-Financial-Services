```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv(r"C:\Users\ZAK-TECH\Desktop\KAIM_week7\data\complaints.csv")
print(df.shape)
df.head()
```

```python
df.info()
df.isnull().sum().sort_values(ascending=False)
```

```python
columns_of_interest = [
    "complaint_id",
    "product",
    "issue",
    "sub_issue",
    "company",
    "state",
    "date_received",
    "consumer_complaint_narrative"
]

df = df[columns_of_interest]
```

```python


df["product"].value_counts().head(10).plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Products by Complaint Volume")
plt.ylabel("Number of Complaints")
plt.show()
```

```python
df["consumer_complaint_narrative"].isna().mean()
```

```python
df = df.dropna(subset=["consumer_complaint_narrative"])
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r"\n+", " ", text)
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^a-zA-Z0-9\s]", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

```python
df["clean_narrative"] = df["consumer_complaint_narrative"].apply(clean_text)
```

```python
df["text_length"] = df["clean_narrative"].apply(len)

df["text_length"].describe()
```

```python
df["text_length"].hist(bins=50, figsize=(10,5))
plt.title("Complaint Narrative Length Distribution")
plt.show()
```

```python
target_products = [
    "Credit card",
    "Personal loan",
    "Savings account",
    "Money transfer"
]

df = df[df["product"].isin(target_products)]
```

```python
sample_df = (
    df.groupby("product", group_keys=False)
    .apply(lambda x: x.sample(min(len(x), 3000), random_state=42))
)

sample_df.shape
```

```python
df.to_parquet("data/cfpb_cleaned.parquet", index=False)
```
